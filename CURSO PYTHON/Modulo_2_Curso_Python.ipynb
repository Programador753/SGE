# Principales tipos de datos


En cualquier lenguaje de programaci√≥n de alto nivel se manejan tipos de datos. Los tipos de datos definen un conjunto de valores que tienen una serie de caracter√≠sticas y propiedades determinadas.

Pensemos por un momento cuando √©ramos j√≥venes üôÑ y est√°bamos en el colegio en clase de matem√°ticas. Seguro que diste alguna clase en la que te ense√±aban los distintos conjuntos de n√∫meros. Los naturales (1, 2, 3, 4, ‚Ä¶), los enteros (‚Ä¶, -2, -1, 0, 1, 2, ‚Ä¶), los reales (‚Ä¶ -1.1, -0.3, 2.1, ‚Ä¶), etc. Pues bien, en programaci√≥n (y por supuesto en Python), cada uno de esos conjuntos ser√≠a lo que llamamos tipo de datos.

En Python, todo valor que pueda ser asignado a una variable tiene asociado un tipo de dato y, este tipo de dato establece qu√© valores puede tomar una variable y qu√© operaciones se pueden realizar sobre la misma.

##¬†**Tipos num√©ricos**

Python define tres tipos de datos num√©ricos b√°sicos: enteros, n√∫meros de punto flotante (simular√≠a el conjunto de los n√∫meros reales, pero ya veremos que no es as√≠ del todo) y los n√∫meros complejos.

**N√∫meros enteros**

El tipo de los n√∫meros enteros es int. Este tipo de dato comprende el conjunto de todos los n√∫meros enteros, pero como dicho conjunto es infinito, en Python el conjunto est√° limitado realmente por la capacidad de la memoria disponible. No hay un l√≠mite de representaci√≥n impuesto por el lenguaje.

Pero tranquilidad, que para el 99% de los programas que desarrolles tendr√°s suficiente con el subconjunto que puedes representar.

Un n√∫mero de tipo int se crea a partir de un literal que represente un n√∫mero entero o bien como resultado de una expresi√≥n o una llamada a una funci√≥n.

```python
a = -1  # a es de tipo int y su valor es -1
b = a + 2  # b es de tipo int y su valor es 1
print(b)
```

Tambi√©n podemos representar los n√∫meros enteros en formato binario o hexadecimal.

Para crear un n√∫mero entero en hexadecimal, hay que anteponer 0x a una secuencia de d√≠gitos en hexadecimal (del 0 al 9 y de la A la F).

En cuanto a los n√∫meros en binario, se antepone 0b a una secuencia de d√≠gitos en binario (0 y 1).

```python
diez = 10
diez_binario = 0b1010
diez_hex = 0xa
print(diez)
print(diez_binario)
print(diez_hex)
```

**N√∫meros de punto flotante**

Vamos a hacer un experimento que te va a dejar a cuadros. Ejecuta el siguiente c√≥digo python:

`````python
resultado = 1.1 + 2.2
print(resultado)
`````



üò≥ ¬øEn serio? üò≥ ¬øQu√© es ese 3 del final?

Al igual que ocurre con los n√∫meros enteros, los n√∫meros reales son infinitos y, por tanto, es imposible representar todo el conjunto de n√∫meros reales con un ordenador.

Para representar el mayor n√∫mero posible de los n√∫meros reales con las limitaciones de memoria, se adopt√≥ la notaci√≥n cient√≠fica de representaci√≥n de n√∫meros reales al sistema binario (que es el sistema que se utiliza en programaci√≥n para representar los datos e instrucciones).

En esta notaci√≥n cient√≠fica, los n√∫meros se representan as√≠:

N√∫mero	Notaci√≥n cient√≠fica
101,1	= 1,011 * 10^2
0,032	= 3,2 * 10^-2

Pero hay m√°s casos, como por ejemplo la representaci√≥n del n√∫mero 1/3.

```python
resultado = 1 / 3
print(resultado)
```

En alg√∫n momento, el ordenador tiene que truncar el n√∫mero peri√≥dico resultante.

La explicaci√≥n final es que los n√∫meros de punto flotante se representan en el hardware del ordenador como fracciones de base 2 (binarias). Y el problema est√° en que la mayor√≠a de las fracciones decimales no se pueden representar de forma exacta como fracciones binarias porque tienen infinitos n√∫meros decimales. Una consecuencia es que, en general, los n√∫meros decimales de punto flotante que usas en tus aplicaciones son una aproximaci√≥n de los n√∫meros binarios de punto flotante realmente almacenados en la m√°quina.

En Python, puedes usar el tipo float sin problemas para representar cualquier n√∫mero real (siempre teniendo en cuenta que es una aproximaci√≥n lo m√°s precisa posible). Por tanto para longitudes, pesos, frecuencias, ‚Ä¶, en los que pr√°cticamente es lo mismo 3,3 que 3,3000000000000003 el tipo float es el m√°s apropiado.

Cuando un n√∫mero float vaya a ser usado por una persona, en lugar de por el ordenador, puedes darle formato al n√∫mero de la siguiente manera:


`````python
real = 1.1 + 2.2  # real es un float
print(real) # 3.3000000000000003 - Representaci√≥n aproximada de 3.3
print(f'{real:.2f}') # 3.30  real mostrando √∫nicamente 2 cifras decimales
`````

Al igual que los n√∫meros enteros, un float se crea a partir de un literal, o bien como resultado de una expresi√≥n o una funci√≥n.

`````python
un_real = 1.1  # El literal debe incluir el car√°cter .
otro_real = 1/2  # El resultado de 1/2 es un float
not_cient = 1.23E3  # float con notaci√≥n cient√≠fica (1230.0)
print(un_real)
print(otro_real)
print(not_cient)
`````

Si por cualquier motivo s√≠ que necesitas una mayor precisi√≥n a la hora de trabajar con los n√∫meros reales, Python tiene otros tipos de datos, como Decimal.

El tipo Decimal es ideal a la hora de trabajar, por ejemplo, con dinero o tipos de inter√©s. Este tipo de dato trunca la parte decimal del n√∫mero para ser m√°s preciso, pero no es el objetivo de este tutorial hablar sobre el tipo de dato Decimal.

# Concepto de variable y convenciones de nombres de variables


Para poder realizar tareas es necesario guardar en la memoria del ordenador los datos que necesita la aplicaci√≥n para funcionar. Por ejemplo, en este momento tu navegador web sabe cu√°ntas pesta√±as hay abiertas y qu√© p√°gina se abri√≥ en cada una, sabe en qu√© posici√≥n de la p√°gina est√°s y qu√© usuario utilizaste para entrar a Google Colab. Todos estos datos est√°n almacenados en la memoria del ordenador y el navegador lee estos datos y los modifica cada vez que lo requiere.
\
\
En programaci√≥n, se le asigna un nombre a cada peque√±o segmento de la memoria que almacena un dato o un valor. De esta forma en el programa cada vez que se usa este nombre se esta refiriendo al valor del dato almacenado en la memoria. Este es el concepto de "variable".
\
\
En python esto se logra de la siguiente manera:

`````python
x = 5
print(x)
`````

Este ejemplo utilizar√° un bloque de la memoria para almacenar el n√∫mero `5`, luego le asigna a este bloque el nombre `x` y por lo tanto a partir de entonces cada vez que se utiliza la palabra `x` en realidad nos referimos al valor almacenado dentro del bloque de memoria, al cual apodamos `x`.

**¬øC√≥mo se interpreta una expresi√≥n de asignaci√≥n?**

Es muy importante entender c√≥mo leer una expresi√≥n de asignaci√≥n.
Cuando recorras y analices tu c√≥digo, es preferible que entiendas el significado de la operaci√≥n:

*nombre* = *expresi√≥n*

Como:
* Se **asigna** el valor de *expresi√≥n* a *nombre*.

Y no como:
* ~~*variable* es *igual* a *expresi√≥n*~~.

**Ejemplos**

En estas variables podemos guardar lo que sea necesario para ejecutar el programa, por ejemplo si necesitamos guardar texto se indica con: "(texto)" o '(texto)'.

`````python
x = 'Hola'
y = "Adios"
z = 123
print(x)
print(y)
print(z)
`````

Ten en cuenta que la instrucci√≥n **print( )** permite mostrar el contenido de una variable. Cuando ponemos **print(x)** no imprime la letra *x* sino que imprime el contenido de **x**. Es por esto que para eliminar la confusi√≥n se usan comillas cuando queremos mostrar texto y no el contenido de una variable.

Es importante tener en cuenta que no es necesario aclarar de antemano si las variables son textos (Strings) o n√∫meros, el lenguaje lo puede identificar autom√°ticamente

**Reglas para el nombre de una variable:**

- Debe empezar con una letra (a-z) o un gui√≥n bajo (_).
- Los otros caracteres pueden ser letras, numeros o gui√≥n bajo.
- Distingue may√∫sculas de min√∫sculas.
- Existen palabras reservadas que no se pueden usar como nombre de variable porque Python las usa para otras cosas. Por ejemplo el nombre de una variable no puede ser print.

# Operaciones b√°sicas con variables


### Operaciones con n√∫meros

No ser√≠a muy √∫til un programa si lo √∫nico que podemos hacer es guardar en una variable y mostrarla, por lo que existen algunas operaciones b√°sicas que nos permiten modificar el contenido:

- Sumar dos variables ( $a+b$ ):

  `a + b`

- Restar dos variables ( $a-b$ ):

  `a ‚àí b`

- Multiplicar dos variables ( $a\cdot b$ ):

  `a * b`

- Dividir una variable por la otra ( $a\div b$ ):

  `a / b`

- Una variable elevada a la otra ( $a^b$ ):

  `a ** b`

- Asignarle un valor a una variable ( $a\leftarrow b$ ):

  `a = b`
  

¬øCual ser√° la salida del siguiente programa?


`````python
x = 3
y = -2
y = x*y
x = x**2
print("y =", y)
print("x =", x)
`````

**¬°Importante!** Tenemos varias formas de mostrar una variable utilizando la funci√≥n print(). Como ya hemos comentado antes, la programaci√≥n es una tarea grupal en la que, de forma simult√°nea, trabajan varias personas. No suele ser buena idea mostrar √∫nicamente el valor de una variable en un print(), siempre es recomendable que esta vaya junto a una frase o palabra que la identifique.

`````python
x = 2
y = 3
print(x) # ERROR
print("x =", x)
print("x =", x, "y =", y)
print(f"x = {x} y = {y}") # M√°s moderno y recomendable (no hay que hacer casting)
`````

Python nos permite ahorrar c√≥digo combinando el operando **=** con alguno de los anteriormente vistos:

*   Suma a la varaible ( $a \leftarrow a+ b$ ):

  `a += b`

*   Resta a la variable ( $a \leftarrow a- b$ ):

  `a -= b`

*   Multiplica a la variable ( $a \leftarrow a\cdot b$ ):

  `a *= b`

*   Divide a la variable ( $a \leftarrow a\div b$ ):

  `a /= b`

¬øCual ser√° la salida del siguiente codigo?

`````python
x = 17
x += 3
x /= 10
print("x =", x)
`````

### Operaciones con cadenas de caract√©res

Una cadena de caracteres en un objeto que consiste precisamente en una serie de signos o caracteres. Una forma de manipular cadenas de caracteres es utilizar operadores de cadenas de caracteres. Dichos operadores se representan con s√≠mbolos que asociamos a las matem√°ticas, como +, -, *, / y =. Estos signos realizan acciones similares a sus contrapartes matem√°ticas cuando se usan con las cadenas de carateres, aunque no iguales.

**Concatenar**

Este t√©rmino significa juntar cadenas de caracteres. El proceso de concatenaci√≥n se realiza mediante el operador de suma (+). Ten en cuenta que debes marcar expl√≠citamente d√≥nde quieres los espacios en blanco y colocarlos entre comillas.

En este ejemplo, la cadena de caracteres ‚Äúmensaje1‚Äù tiene el contenido ‚ÄúHola Mundo‚Äù

`````python
mensaje = 'Hola' + ' ' + 'Mundo'
print(mensaje)

mensaje = 'Hola'
mensaje3 = 'Mundo'
mensaje_final = mensaje + mensaje3
print(mensaje_final)
`````

**Multiplicar**

Si quieres varias copias de una cadena de caracteres utiliza el operador de multiplicaci√≥n (*). En este ejemplo, la cadena de caracteres mensaje2a lleva el contenido ‚ÄúHola‚Äù tres veces, mientras que la cadena de caracteres mensaje2b tiene el contenido ‚ÄúMundo‚Äù.

`````python
mensaje_2a = 'Hola ' * 3
mensaje_2b = 'Mundo'
print(mensaje_2a + mensaje_2b)
`````

# Control de Flujo: Estructura IF y bucles


Los programas siguen un "flujo" definido por el programador. Hasta ahora el flujo del programa siempre fue uno s√≥lo, se ejecuta una instrucci√≥n, luego la siguiente, luego la siguiente a esta, etc. Sin embargo es posible realizar tareas diferentes dependiendo de ciertas condiciones, repetir segmentos de c√≥digo, y modificar el flujo del programa de forma que no sea simplemente una secuencia lineal de instrucciones. Esto permite realizar tareas mucho m√°s complejas, en la cu√°l el programa tome decisiones respecto a lo que debe hacer en funci√≥n de la situaci√≥n.

**Nota:** estas estructuras terminan con un ":" , y la siguiente l√≠nea tiene un tab (como una sangr√≠a).

### Estructura if

El if es la estructura de control de flujo b√°sica que le permite al programador analizar si se cumple cierta condici√≥n.

Se realizar√°n distintas acciones, seg√∫n si la condici√≥n se cumple o no.

Las distintas formas de implementar la estructura son las siguientes:
```python
if (condicion):
    #c√≥digo que se ejecuta si se cumple la condici√≥n
```
```python
if (condicion):
    #c√≥digo que se ejecuta si se cumple la condici√≥n
else:
    #c√≥digo que se ejecuta si no se cumple la condici√≥n
```
```python
if (condicion1):
    #c√≥digo que se ejecuta si se cumple la condici√≥n1
elif (condicion2):
    #c√≥digo que se ejecuta no se cumple la condicion1, pero s√≠ la condicion2
elif (condicion3):
    #c√≥digo que se ejecuta no se cumple la condicion1 ni la condici√≥n2, pero s√≠ la condici√≥n3
else:
    #c√≥digo que se ejecuta no se cumple ninguna condici√≥n
````

Se debe prestar especial atenci√≥n a la indentaci√≥n/tabulaci√≥n. Siempre que se define un if se crea un bloque de c√≥digo (un conjunto de instrucciones) que son las que se ejecutan cuando se cumple la condici√≥n dada. Este bloque de c√≥digo debe tener una "sangr√≠a", de esta forma Python comprende d√≥nde empieza y d√≥nde termina este bloque. Si se utiliza la "sangr√≠a" incorrectamente, Python indicar√° que hay un error en el c√≥digo y no se podr√° ejecutar.

Siempre se comienza con if, despu√©s puede haber cualquier cantidad de elif, como ninguno, y, finalmente, se puede terminar con else, como no hacerlo. Luego del else no puede haber m√°s elif ni else.

**Nota:** Python es un lenguaje de programaci√≥n TABULADO. Es decir, las instrucciones que van dentro de un bloque de instrucciones est√°n tabuladas para que el ordenador las reconozca. Esto no ocurre en otros lenguajes de programaci√≥n como Java (utiliza llaves {}). Esto hace que en Python, la importancia de tabular el c√≥digo de forma correcta sea todav√≠a mayor. No solo por legibilidad, si no tambi√©n por funcionamiento.

**Nota2:** ¬øPor qu√© usar elif y no usar if? Hay peque√±as diferencias entre ellos. Un bloque condicional √∫nicamente tiene que tener un solo IF, en el momento que ponemos 2 IFs seguidos estamos hablando de 2 bloques condicionales distintos.

El if siempre se usa para evaluar la primera condici√≥n. Si esta es True, el bloque de c√≥digo dentro de if se ejecuta y el resto de las condiciones se ignoran.

El elif se usa despu√©s de un if para evaluar condiciones adicionales solo si la primera no se cumple:
* Se pueden usar m√∫ltiples elif en una estructura.
* Si un elif se cumple, los siguientes elif (y el else si lo hay) se ignoran.

**¬øQu√© pasa si solo usamos varios if?**
Cada if se eval√∫a independientemente, por lo que pueden ejecutarse varios bloques a la vez si sus condiciones son True.

`````python
x = 3
if (x > 1):
  print("Hola")
elif (x == 3):
  print("Adios")
else:
  print("Hasta Luego")
`````

## Estructura while

La estructura while permite implementar ciclos en nuestro programa (tambi√©n llamados loops o bucles). Es decir, repetir una porci√≥n del c√≥digo una y otra vez siempre que se cumpla la condici√≥n. La estructura b√°sica es la siguiente:

```python
while (condicion):
    #c√≥digo que se ejecuta en tanto se cumpla la condicion
    #incluir una operaci√≥n que modifique la condici√≥n
```

$\color{red}{\text{Cuidado:}}$ Es importante incluir una operaci√≥n que modifique la condici√≥n porque, sino, el loop se ejecutar√° infinitamente (o hasta que el programa finalice de forma abrupta).


`````python
x = 1
while(x < 10):
    print(x)
    x = x + 1

print('Termina el while')
print("Continuo con el resto del programa")
`````

Es importante tener en cuenta que la condici√≥n del **while** se eval√∫a s√≥lo al principio de cada iteraci√≥n y no se volver√° a evaluar hasta no terminar de ejecutar todas las instrucciones contenidas en el bloque.

Al usar una variable auxiliar para contar la cantidad de repeticiones del **while**, es recomendable no modificar su valor dentro del bloque excepto para incrementar/decrementar su valor acordemente. Esto es porque se corre el riesgo de que la condici√≥n siempre sea *True* y el **while** se ejecute infinitamente.

Adem√°s, es dificil darse cuenta de que la cantidad de iteraciones ser√° de hecho la que queremos.

¬øQu√© realiza este programa?

`````python
nota = int( input('Ingrese un n√∫mero del 1 al 10: ') )

while nota < 1 or nota > 10:
  print('Fuera de rango!')
  nota = int( input('Ingrese un n√∫mero del 1 al 10: ') )

print("La nota es mayor a 4?", nota >= 4 )
`````

## Estructura for

La estructura del **for** permite realizar una serie de acciones con una variable, que cambia su valor en cada repetici√≥n (tambi√©n llamada *iteraci√≥n*).

```python
for (variable) in (conjunto de valores):
    #c√≥digo que se ejecuta para cada elemento en el conjunto de valores
```

Ve√°moslo en un ejemplo de c√≥digo:

`````python
for x in range(0,15,5):
    print(x)
`````

La instrucci√≥n `range(a,b,d)` genera los valores desde *a* hasta *b* (sin incluir) con incrementos de *d*.

Por ejemplo, el siguiente codigo generar√° los numeros pares entre 0 y 8:

`````python
init = 0
end = 8 + 2
step = 2

for x in range(init,end,step):
    print(x)
`````

Al utilizar la instrucci√≥n **range( )** no es necesario proporcionar siempre los 3 par√°metros. Observen qu√© sucede cuando se proporcionan solamente 1 o 2 par√°metros:

`````python
print("range(5):")
for x in range(5):
    print(x)

print("range(10, 15):")
for x in range(10, 15):
    print(x)
`````

## Instrucciones break y continue

Usar bucles for y bucles while en Python le permite automatizar y repetir tareas de manera eficiente.

Sin embargo, a veces, es posible que un factor externo influya en la forma en que se ejecuta su programa. Cuando esto sucede, es posible que prefiera que su programa cierre un bucle por completo, omita parte de un bucle antes de continuar o ignore ese factor externo. Puede hacer estas acciones con las instrucciones break y continue.

**Instrucci√≥n break**

En Python, la instrucci√≥n break le proporciona la oportunidad de cerrar un bucle cuando se activa una condici√≥n externa. Debe poner la instrucci√≥n break dentro del bloque de c√≥digo bajo la instrucci√≥n de su bucle, generalmente despu√©s de una instrucci√≥n if condicional.

Veamos un ejemplo en el que se utiliza la instrucci√≥n break en un bucle for:

`````python
number = 0

for number in range(10):
    if number == 5:
        break    # break here
    print('Number is ' + str(number))

print('Out of loop')
`````

**Instrucci√≥n continue**

La instrucci√≥n continue da la opci√≥n de omitir la parte de un bucle en la que se activa una condici√≥n externa, pero continuar para completar el resto del bucle. Es decir, la iteraci√≥n actual del bucle se interrumpir√°, pero el programa volver√° a la parte superior del bucle.

La instrucci√≥n continue se encuentra dentro del bloque de c√≥digo abajo de la instrucci√≥n del bucle, generalmente despu√©s de una instrucci√≥n if condicional.

Usando el mismo programa de bucle for que en la secci√≥n anterior Instrucci√≥n break, emplearemos la instrucci√≥n continue en vez de la instrucci√≥n break:

`````python
number = 0

for number in range(10):
    if number == 5:
        continue    # continue here

    print('Number is ' + str(number))

print('Out of loop')
`````

`````python
numero1 = 0
numero2 = 1

print(numero1)
print(numero2)

iteraciones = 0
# Solo 8 veces porque los dos primeros n√∫meros los imprimimos fuera
while iteraciones < 8:
  nuevo_numero = numero1 + numero2
  numero1 = numero2
  numero2 = nuevo_numero
  print(nuevo_numero)

  iteraciones += 1
`````
