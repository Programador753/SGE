## Request

El módulo requests es una biblioteca de Python utilizada para hacer peticiones HTTP de manera sencilla. Nos permite interactuar con APIs, descargar contenido de páginas web y enviar datos a servidores de forma rápida y legible.

```
import requests

response = requests.get("https://www.filmaffinity.com/es/film809297.html")  # Hacemos la petición
print(response.status_code)  # Imprime el código de estado de la respuesta
print(response.text)  # Convierte la respuesta en un diccionario de Python
```

Ya tenemos la etiqueta seleccionada... ahora solo nos queda utilizar alguna de las funciones que ya conocemos de los String, a modo de juego, para intentar sacar el contenido que queremos:

```
import requests

response = requests.get("https://www.filmaffinity.com/es/film809297.html")  # Hacemos la petición
print(response.status_code)  # Imprime el código de estado de la respuesta
print(response.text.splitlines())  # Convierte la respuesta en un diccionario de Python

for etiqueta in response.text.splitlines():
  if "sinopsis:" in etiqueta.lower() and "<meta name" in etiqueta.lower():
    print(etiqueta)
```

```
import requests

response = requests.get("https://www.filmaffinity.com/es/film809297.html")  # Hacemos la petición
print(response.status_code)  # Imprime el código de estado de la respuesta
print(response.text.splitlines())  # Convierte la respuesta en un diccionario de Python

for etiqueta in response.text.splitlines():
  if "sinopsis:" in etiqueta.lower() and "<meta name" in etiqueta.lower():
    print(etiqueta.lower().split("sinopsis:")[1])
```

Si queremos ser más exactos, podemos utilizar las etiquetas html para acceder al contenido que queremos:

```
import requests

response = requests.get("https://www.filmaffinity.com/es/film809297.html")  # Hacemos la petición

for etiqueta in response.text.splitlines():
  if "<dd class=\"\" itemprop=\"description\">" in etiqueta.lower():
    print(etiqueta)
```

Ahora solamente nos falta limpiar la información. Podemos utilizar la función replace() de los strings:

```
import requests

response = requests.get("https://www.filmaffinity.com/es/film809297.html")  # Hacemos la petición

for etiqueta in response.text.splitlines():
  if "<dd class=\"\" itemprop=\"description\">" in etiqueta.lower():
    print(etiqueta.replace("<dd class=\"\" itemprop=\"description\">", "").replace("(FILMAFFINITY)</dd>", "").strip())
```

**Nota:** existe una herramienta muy útil que se utiliza en programación para buscar o eliminar patrones dentro de un gran volúmen de datos de forma óptima tanto en tiempo como en memoria. Esta herramienta se llama **Regex** o Expresiones Regulares. Se adjunta en el siguiente [enlace](https://www.w3schools.com/python/python_regex.asp) la documentación de Regex en Python.

No vamos a profundizar mucho en esta herramienta pero si que vamos a ver un pequeño ejemplo de su funcionamiento:

```
import requests
import re

response = requests.get("https://www.filmaffinity.com/es/film809297.html")  # Hacemos la petición

for etiqueta in response.text.splitlines():
  if "<dd class=\"\" itemprop=\"description\">" in etiqueta.lower():
    print(re.sub(r"<[^>]+>", "", etiqueta).strip())
```

# PIP

Un **gestor de paquetes** es una herramienta que automatiza el proceso de instalación, actualización, configuración y eliminación de paquetes de manera coherente.  

Los gestores de paquetes están diseñados para eliminar la necesidad de realizar instalaciones y actualizaciones manuales, garantizando así que un paquete se instale junto con todas las dependencias que necesita para funcionar. Además, dado que los gestores de paquetes aprovechan la información almacenada en repositorios de paquetes certificados, como **PyPi** y **Anaconda**, aseguran la integridad y autenticidad de los paquetes.  

El gestor de paquetes más popular para **Python** es **pip**. Desarrollado en 2008, **pip** (acrónimo de *"pip Install Packages"*) es hoy en día la herramienta estándar para instalar paquetes de Python y sus dependencias de manera segura. La mayoría de las versiones recientes de Python incluyen **pip** preinstalado. **Python 2.7.9 y Python 3.4 (y versiones posteriores) incluyen pip por defecto.**  

**Pip** es una herramienta poderosa y fácil de usar que permite gestionar paquetes de Python con unos pocos comandos. Aunque **pip** utiliza **PyPi** como repositorio predeterminado para obtener paquetes, también tiene la capacidad de instalar paquetes desde otras fuentes, incluyendo:  

- **Sistemas de control de versiones** como GitHub, Mercurial, Subversion y Bazaar.  
- **Archivos de requisitos**. Normalmente, los paquetes de Python dependen de varios otros paquetes para ejecutarse. Para instalar todas las dependencias necesarias a la vez, **pip** utiliza el archivo `requirements.txt`, que contiene una lista de los paquetes necesarios junto con sus versiones correctas.  
- **Archivos de distribución**. Estos son archivos versionados y listos para instalar que contienen paquetes de Python, módulos y otros archivos de recursos necesarios para que un paquete funcione. Se presentan en dos formas:  
  - **Distribución de código fuente** (generalmente abreviado como *"sdist"*)  
  - **Distribución en formato Wheel** (abreviado como *"wheel"*)  

Para usar **pip** y comenzar a administrar paquetes, primero debes asegurarte de que esté instalado en tu ordenador. Para verificar si **pip** está disponible, ejecuta uno de los siguientes comandos en la línea de comandos:

```
#pip3 --version
!pip3 --version
```

```
#pip --version
!pip --version
```

Después de leer la sección anterior, es posible que te preguntes cuál es la diferencia entre **pip** y **pip3**. Tras el lanzamiento de **Python 3**, **pip** incorporó el nuevo comando **pip3**, que siempre opera en el entorno de **Python 3** de tu ordenador. Lo mismo ocurre con el comando **pip2** para **Python 2**. Por lo tanto, si quieres asegurarte de que **pip** opere en tu entorno de **Python 3** o **Python 2**, usa los comandos **pip3** o **pip2**, respectivamente.  

Por otro lado, el comando **pip** funciona en el entorno de Python que sea apropiado según el contexto. Esto es relevante cuando tienes instaladas tanto la versión **Python 2** como **Python 3** en tu ordenador.  

## Instalador paquetes con PIP




El uso más común de **pip** es la instalación de paquetes de Python. Por ejemplo, si quieres instalar **pandas**, el paquete estándar para manipular data frames en Python, la forma más sencilla de hacerlo es ejecutando la siguiente instrucción:

```
!pip install pandas
```

Puedes necesitar instalar una versión específica, en ese caso deberías usar el siguiente comando:

```
!pip install pandas==1.4.0
```

O también puede darse el caso de que necesites cualquier versión que se encuentre entre un rango inicial de versiones. Para ello:

```
!pip install pandas>=1.0.0,<1.5.0
```

### Instalar una lista de paquetes usando el fichero requisitos

Cuando trabajas en proyectos colaborativos, es muy común que todos los miembros del equipo utilicen los mismos paquetes con las mismas versiones. Para garantizar esto, la mejor manera es instalar los paquetes utilizando un archivo de requisitos. Este suele ser un archivo de texto que contiene todos los paquetes, junto con sus respectivas versiones, que se usan en el proyecto.

Pip permite instalar una lista de paquetes de una sola vez utilizando un archivo de requisitos. Por ejemplo, si necesitamos los paquetes **numpy**, **pandas** y **TensorFlow** para nuestro proyecto, podríamos incluirlos, junto con las versiones deseadas, en un archivo **requirements.txt**, como se muestra a continuación:

```
!pip install -r /content/requirements.txt
```

Si queremos generar rápidamente un fichero requirements.txt para compartirlo con nuestros miembros del equipo, podemos hacerlo de la siguiente manera:

```
!pip freeze > /content/requirements.txt
```

## Actualizar paquetes con PIP

A veces, necesitarás actualizar a una versión más reciente de un paquete que ya tienes instalado en tu ordenador. **Pip** hace que este proceso sea extremadamente fácil. Por ejemplo, si quieres actualizar **pandas** a la última versión:

```
!pip install --upgrade pandas
```

## Eliminar paquetes con PIP

Eliminar un paquete con PIP es extremadamente sencillo. Simplemente hay que utilizar el siguiente comando:

```
!pip uninstall pandas
```

## Otras funciones de PIP

Si necesitamos información sobre el comando, para saber como utilizarlo, podemos acceder a su documentación usando:

```
!pip help
```

Si queremos listar todos los paquetes que tenemos instalados en nuestro entorno de trabajo:

```
!pip list
```

Si quieres información sobre un pquete en específico antes de instalarlo:

```
!pip show pandas
```
