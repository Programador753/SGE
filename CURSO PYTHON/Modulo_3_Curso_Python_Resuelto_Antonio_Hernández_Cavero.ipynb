## **1. Definici√≥n y uso de funciones**

Podemos pensar que una funci√≥n es como tomar un conjunto de instrucciones y empaquetarlas bajo el mismo nombre, es como un peque√±o programa dentro del programa, una manera de crear nuestras propias instrucciones. Las funciones son bloques que nos permiten acortar el c√≥digo, no repitiendo estructuras que necesitamos utilizar mucho. Al igual que los programas las funciones tienen un *input* y un *output*: el input son las variables de entrada y su output es el valor o valores que devuelven.

En Python las funciones se implementan de la siguiente forma:

```python
def miFuncion(a, b, c, ...):
    
    #Hago lo que necesite con las variables a, b, c, ...
    
    return x , y , z , ...
```
Aqu√≠, el nombre de la funci√≥n es `miFuncion`, los par√°metros de entrada son las variables `a , b , c , ...` que se indican entre par√©ntesis, y los par√°metros que la funci√≥n devuelve son `x , y , z , ...` que se indican con la instrucci√≥n **return**.

Por ejemplo, supongamos que frecuentemente necesito calcular las ganancias totales generadas por cierto producto que se encuentra a la venta en mi negocio. Cada vez que quiera realizar este c√°lculo tendr√© que obtener el valor de cada venta, restarle el costo de fabricaci√≥n y multiplicarlo por la cantidad de ventas. Usando una funci√≥n puedo crear mi propia instrucci√≥n a la que se le ingrese el c√≥digo de producto y devuelva el resultado. As√≠ puedo escribir el nombre de la funci√≥n cada vez que quiera hacer este c√°lculo sin tener que escribir nuevamente el procedimiento completo.

```python
def miFuncion(producto): # recibo el codigo del producto
    cantidad_ventas, valor_venta, costo_fabricacion = buscar_producto(producto)
    return cantidad_ventas*(valor_venta - costo_fabricacion)
```

En el caso de que la funci√≥n no necesite recibir informaci√≥n, los par√©ntesis se deben dejar vac√≠os. Al ejecutar la funci√≥n no hay que olvidarse de escribir estos par√©ntesis, incluso aunque est√©n vac√≠os. Es la forma que tiene Python de diferenciar variables de funciones:

````python
def saludar():
    print( "¬°Hola! Que tengas un buen d√≠a :)")

print("A continuaci√≥n se imprimir√°n tres saludos:")
saludar()
saludar()
saludar()
````

````python
def saludar_devolucion():
  mensaje = "¬°Hola! Que tengas un buen d√≠a :)"
  return mensaje

print(saludar_devolucion())
````

**Nota:** no siempre es necesario poner la instrucci√≥n return al final de la funci√≥n. En este caso la funci√≥n termina cuando finaliza el bloque indentado (con sangr√≠a).

Es importante tener en cuenta que la definici√≥n de la funci√≥n no ejecutar√°n el c√≥digo de las mismas. Solamente le indica a la computadora que, cuando se escriba el nombre de la funci√≥n, esta debe realizar las instrucciones indicadas por la definici√≥n.

**Nota 2:** Es importante saber que Python es un lenguaje interpretado y ejecuta el c√≥digo l√≠nea por l√≠nea. Esto significa que si intentas llamar a una funci√≥n antes de declararla, obtendr√°s un error.

````python
saludar()  # ‚ùå Error: NameError: name 'saludar' is not defined

def saludar():
    print("¬°Hola!")
````

Aqu√≠, el int√©rprete a√∫n no ha le√≠do la definici√≥n de saludar(), as√≠ que cuando intentas llamarla antes, no sabe qu√© es.

Otros lenguajes como Java, son lenguajes compilados y tienen un proceso de compilaci√≥n previo a la ejecuci√≥n. Durante la compilaci√≥n, el compilador escanea todo el c√≥digo y ya conoce todas las funciones antes de que se ejecuten. Por lo tanto, en estos lenguajes, no es necesario crear las funciones antes de llamarlas.

Veamos ahora un ejemplo de funci√≥n:

````python
def suma(a, b):
    s = a + b
    return s

z = suma(3, 4)
print("3 + 4 =", z)

x = int(input("Ingrese el primer sumando: "))
y = int(input("Ingrese el segundo sumando: "))
z = suma(x, y)
print("x + y =", z)
````

En este ejemplo podemos ver que los nombres que le damos a las variables de input (a, b) son utilizados de forma interna a la funci√≥n. Cuando ejecutamos la funci√≥n con los valores 3 y 4, las variables pasan a tomar estos valores y luego se ejecutan las instrucciones correspondientes. Cuando ejecutamos la funci√≥n con los par√°metros x, y lo que sucede es que a y b copian los valores de estas variables indicadas.

La instrucci√≥n return sirve para indicar el resultado que la funci√≥n "devuelve". Una vez que se alcanza la instrucci√≥n return la funci√≥n termina y devuelve el valor indicado. En general es preferible devolver informaci√≥n usando return antes que imprimirla usando print ya que esto permite reutilizar la funci√≥n para una cantidad de casos mayor, donde no siempre quiero mostrar el resultado en pantalla, aunque queda a criterio de cada uno para lo que necesite.

Al concepto de funci√≥n tambi√©n podemos combinarlo con las estructuras que vimos previamente:

````python
def chequarContrase√±a(c):
    if c == "Secreto":
        resultado = True
    else:
        resultado = False
    return resultado


ingresado = input("Ingrese contrase√±a: ")
while not chequarContrase√±a(ingresado):
    ingresado = input("Contrase√±a incorrecta. Ingrese de nuevo:")
else:
    print("Contrase√±a correcta.")
````

### Listas

Las listas nos permiten almacenar elementos de manera secuencial, los cuales pueden ser accedidos mediante el conocimiento de su posici√≥n en la lista. Estas las creamos usando corchetes de la siguiente manera:

> a = **[** $ a_{0}, a_{1}, a_{2}, \dots $ **]**

A continuaci√≥n se mostrara un ejemplo sencillo con listas:

````python
# Creamos una lista con elementos
a = [ 13, 40, 10, 30, 67 ]
print(a)

# Creamos una lista vac√≠a
b = []
print(b)

# Creamos una lista vac√≠a
c = list()
print(c)
````

Para acceder a un elemento de una lista se utiliza `lista[√≠ndice]` de forma similar a como funcionan los strings.
Podemos modificar el elemento asign√°ndole un valor.

Al igual que los strings, las listas permiten hacer **slicing**. Esto significa obtener una sub-lista a partir de valores de comienzo, fin y step indicados entre corchetes y separados por dos puntos.

````python
a = [5, 6, 7, 8]
print('a =', a)
print('a[0] =', a[0])
print('a[2] =', a[2])

a[2] = 0
print('a =', a)
````

````python
# Se pueden intercambiar dos elementos de una lista de la siguiente forma
a = [5, 6, 7, 8]

a[0], a[1] = a[1], a[0]
print('a =', a)
````

````python
# Se pueden usar signos negativos para referir los √≠ndices desde el final para atras
a = [5, 6, 7, 8]
print('a[-1] =', a[-4])
````

````python
# Slicing
a = [5, 6, 7, 8]
print('a[1:3] =', a[1:3])

# Se puede omitir el parametro de inicio para comenzar desde el principio
# Se puede omitir el parametro de fin para seguir hasta el final
# Se puede usar un salto negativo para recorrer la lista en sentido inverso
# Se puede invertir una lista de la siguiente forma
print('a[ : :-1] =', a[ : :-1] )
````

**Nota:** a diferencia de lo que dir√≠a el sentido com√∫n, el primer elemento de la lista se accede con el √≠ndice $0$. Esto sucede en muchos lenguajes de programaci√≥n hoy en d√≠a y una vez que uno se acostumbra resulta muy pr√°ctico empezar a contar desde $0$ para aplicaciones de programaci√≥n.

Otro tema a tener en cuenta es que no es posible ponerle el nombre *list* a una lista, ya que es una palabra reservada. Esta es una funci√≥n para generar una lista a partir de otro objeto, como un string. Por ejemplo:

````python
texto = "Hola"
lista = list(texto)

print(texto)
print(lista)
````

Una forma alternativa de generar una lista es utilizando la instrucci√≥n *for* dentro de los corchetes de la siguiente manera:

> b = **[** x **for** x **in** **range**(N) **]**

````python
b = [x for x in range(10)]
print(b)

# Este c√≥digo tan dif√≠cil a simple vista, es lo mismo que este:
b = []
for numero in range(10):
  b.append(numero)
print(b)
````


Podemos tambi√©n aplicar una funcion a la variable x
> a = **[** f(x) **for** x **in** **range**(N) **]**

````python
def multiplicarPor2(x):
  return x*2

a = [x for x in range(15)]
print(a)

b = [multiplicarPor2(x) for x in range(15)]
print(b)
````

Tambi√©n es posible utilizar un objeto iterable, en lugar de usar **range( )**.

````python
lista = [1, 2, 3, 4, 5]
for numero in lista:
  print(numero)
````

#### Operaciones sobre una lista
Podemos realizar m√∫ltiples operaciones sobre una lista:

- **a + b**: al igual que los strings se agrega el contenido de **b** al final de **a**.

````python
a = [1,2]
b = [3,4]
z = a + b
print(z)
````

- **len**( $lista$ ): obtiene el largo de una lista.

````python
a = [1,2,3,4]
tamano = len(a)
print(tamano)
````

- **.sort**(): ordena los elementos de la lista.

````python
a = ["Hola", "Adios", "Como esta", "Buen d√≠a" ]
a.sort()
print(a)

b = [1, 2, 3, 123, 23, 12]
b.sort()
print(b)

# Es posible hacer un orden inverso indicando reverse=True
b.sort(reverse=True)
print(b)
````

Es posible definir una regla de ordenamiento personalizada. Se debe definir una funci√≥n, la cu√°l ser√° evaluada en cada elemento de la lista, y el resultado de esta funci√≥n se usa como criterio de orden. En la funci√≥n *sort* se debe ingresar como par√°metro *key=nombre_de_funcion*.

````python
# En este ejemplo la regla es la longitud del elemento
def mi_orden1(e):
  return len(e)

a = ["Hola", "Adios", "Como esta", "Buen d√≠a"]
a.sort(key=mi_orden1)
print("Orden final:", a)

# En este ejemplo se analiza el segundo valor de cada elemento
def mi_orden2(e):
  return e[1]

area = [ ["Argentina", 2.78], ["Brazil", 8.51], ["Mexico", 1.96] ]
area.sort(key=mi_orden2)
print("Orden final:", area)
````

- **.append**($algo$): permite agregar un elemento al final de la lista.

````python
mensaje = "Hola como estas"
v = [50, mensaje, 150.5, True]
print(v)

# Agrego al final de la lista un elemento que vale 200 (tipo int)
v.append(200)
print(v)
````

- **.extend**($lista$): permite agregar elementos de una lista al final de otra. Es muy similar a la suma de listas.

````python
v = [1, 2, 3]
v2 = ["Uno", "Dos", "Tres"]
v.extend(v2)
print(v)
````

- **.pop**(): permite remover el ultimo elemento de la lista.

````python
v = ["Uno", 2, "Tres", 4, "Cinco"]
v.pop() # Quito el ultimo elemento
print(v)
````

- **.remove**($valor$): remueve el primer elemento de la lista cuyo valor sea el indicado.

````python
v = [1,2,1,2]
v.remove(2) # elimino el primer valor igual a 2
print(v)
v.remove(2) # elimino el primer valor igual a 2
print(v)
````

- **del** lista**[*√≠ndice*]**: para eliminar un elemento en una posici√≥n determinada.

````python
a = ["hola","como","estas"]
del a[1]
print(a)
````

- $algo$ **in** lista: muchas veces es necesario saber si un elemento esta dentro de una lista, para ello utilizamos esta herrramienta.


````python
x = [1,2,3,5]
if 5 in x:
    x.remove(5)
else:
  print(5,'no esta en la lista')

print(x)
````

- **max**($lista$): Obtiene el valor m√°s alto de la lista.

````python
valor = max( [1, 2, 3, 4, 5, 4, 3, 2, 1] )
print(valor)

valor = max( ["Curso", "De", "Python"] )
print(valor)
````

- **min**($lista$): Obtiene el valor m√°s bajo de la lista.

````python
valor = min( [1, 2, 3, 4, 5, 4, 3, 2, 1] )
print(valor)

valor = min( ["Curso", "De", "Python"] )
print(valor)
````

**Nota importante:** En Python las listas (list) son mutables, lo que significa que puedes modificar su contenido sin necesidad de crear una nueva lista.



````python
# Crear una lista
a = [1, 2, 3]
b = a  # 'b' apunta a la misma lista que 'a'

# Modificamos la lista a trav√©s de 'b'
b.append(4)

print(a)  # [1, 2, 3, 4]  <-- 'a' tambi√©n se modific√≥
print(b)  # [1, 2, 3, 4]
````

 Aqu√≠ a y b apuntan a la misma lista en memoria, y como las listas son mutables, cualquier cambio hecho en b tambi√©n afecta a a.

 Las listas son mutables, lo que significa que pueden crecer y cambiar de tama√±o. Debido a esto:

* Se asigna un bloque de memoria contiguo inicial.
* Si la lista crece y no hay espacio suficiente, se crea una nueva ubicaci√≥n en memoria m√°s grande y se copian los elementos.
* Cada elemento de la lista es un puntero a su valor real en la memoria.

Adem√°s, hay que tener en cuenta que:
* La lista almacena punteros a los datos reales, por lo que los valores no est√°n necesariamente en posiciones contiguas.
* Si la lista crece y no hay espacio continuo, se mueve toda la lista a una nueva direcci√≥n de memoria.

Vamos a ver unos cuantos ejemplos de todo esto que se acaba de comentar:

````python
lista = [ 1, 2, 3, 4 ]
print(f"Lista = {lista}")

copia = lista
print(f"Copia = {copia}")

copia.remove(1)
print(f"Copia = {copia}")
print(f"Lista = {lista}")
````

¬øC√≥mo hacemos una copia **profunda?**

````python
import copy

lista = [ 1, 2, 3, 4 ]
copia = copy.deepcopy(lista)
copia.remove(1)
print(f"Copia = {copia}")
print(f"Lista = {lista}")

# Otra forma m√°s "normal"
lista = [ 1, 2, 3, 4 ]
copia = []
for num in lista:
  copia.append(num)

copia.remove(1)
print(f"Copia = {copia}")
print(f"Lista = {lista}")
````

Todo parece muy bonito pero... la segunda forma de hacer la copia tiene un peque√±o problema. Como ya sabemos, las listas son punteros a datos que tenemos alojados en la memoria de nuestro ordenador. Si nosotros recorremos una lista
con todo valores num√©ricos, simplemente vamos cogiendo uno a uno con un bucle for y este valor lo escribimos sobre la nueva tabla. ¬øQu√© ocurre si dentro de esta lista, adem√°s de tener n√∫meros, tenemos... **otra lista**? üòß

````python
lista = [ 1, 2, 3, [1, 2, 3] ]
copia = []
for num in lista:
  copia.append(num)
print(f"Copia = {copia}")
print(f"Lista = {lista}")

# Vamos a modificar de la lista interna, el primer dato... cambiando su valor a 6
copia[3][0] = 6
print(f"Copia = {copia}")
print(f"Lista = {lista}")
````

Como podemos ver, en este caso, si que se ha modificado en ambas listas. Cosa que NO queremos. Hay que tener en cuenta qu√©, cuando en el bucle hemos ido recorriendo elemento por elemento, cuando ha llegado a la lista interna, no es capaz de leer los elementos si no que lee una referencia a memoria que apunta al inicio de la lista interna. Como hemos copiado sin saberlo esa referencia, cada vez que hagamos el cambio de una de las lista, se har√° en las dos.

**NOTA IMPORTANTE:** si queremos hacer una copia profunda de una lista, siempre es recomendable usar la llamada a la librer√≠a COPY.

#### Creacion de tuplas

¬øPara que se usan entonces si con las listas puedo hacer lo mismo y mas?

Es una buena pregunta y no hay una respuesta definitiva y clara pero podemos decir que en la practica se utilizan comunmente en los siguientes casos:

1. Cuando queremos devolver mas de un elemento de una funcion.
2. Cuando creamos una lista de elementos de distintos tipos, por ejemplo ("Caramelos",327,2.5)
3. Cuando queremos usar mas de un elemento como clave de un diccionario.

Los casos 2 y 3 los veremos en las siguientes unidades del curso.

Veamos ahora el caso 1. Crearemos una funcion que recibe una lista de numeros y devuelve dos valores, el menor y el mayor de dicha lista:

````python
def menor_y_mayor(lista):
    menor = min(lista)
    mayor = max(lista)
    tupla = (menor, mayor)
    return tupla
````

Veamos como funciona y que devuelve:

````python
x = menor_y_mayor([9,3,6,7,2,8,7])
print(x)
print(x[0])
print(x[1])
````

Hay una forma mas utilizada de utilizar los valores de una lista que se llama unpacking y permite copiar cada elemento de una tupla a una variable distinta, se utiliza de la siguiente forma:

````python
mi_tupla = (100, 200)
elemento1, elemento2 = mi_tupla
print(elemento1)
print(elemento2)

mi_tupla = (100, 200, 300)
elemento1, _, elemento3 = mi_tupla
print(elemento1)
print(elemento2)
````

Si bien el unpacking funciona tambien con listas es mas comun utilizarlo con tuplas.

**Nota importante:** Las tuplas son inmutables, lo que significa que su tama√±o no cambia nunca.

* Python asigna un solo bloque de memoria contiguo para toda la tupla.
* Como no cambia, no hay necesidad de reservar m√°s memoria de la necesaria.
* Al ser m√°s eficiente, las tuplas consumen menos memoria que las listas.

Adem√°s, hay que tener en cuenta que:
* La tupla se almacena en memoria de forma contigua.
* Como la tupla es inmutable, Python puede optimizar el almacenamiento y hacer que su acceso sea m√°s r√°pido.

Bien... ahora esto... se complica un poco m√°s. Hemos dicho que las tuplas son inmutables, y es cierto... Pero, si en ellas almacenamos una estructura de datos mutable (como es el caso de las listas), la cosa cambia...

````python
# Se puede apreciar que Python nos deja acceder a la lista y modificar
# uno de sus elementos. Pero sin embargo, al resto de datos de la tupla no.
tupla = ([1, 2], [3, 4])
tupla[1][0] = 20
tupla[0][1] = 4
print(tupla)

tupla[1] = 16
print(tupla)
````

Con las tuplas nos encontramos con el mismo problema que las listas:

````python
tupla = ([1, 2], 2)
tupla2 = tupla

tupla2[0][1] = 20
print(tupla2)
print(tupla)
````

### Diccionarios

Los estructuras de datos que ha visto hasta ahora (listas y tuplas) usan enteros como √≠ndices; si intenta usar cualquier otro tipo como √≠ndice provocar√° un error.

Los diccionarios son un tipo diferente de colecci√≥n. Son el tipo de mapeo incorporado de Python. Un mapa es una colecci√≥n no ordenada, asociativa. La asociaci√≥n, o asignaci√≥n, es de una clave, que puede ser de cualquier tipo inmutable, a un valor, que puede ser cualquier objeto de datos de Python (mutables o no mutables).

![texto alternativo](https://bjc.edc.org/bjc-r/img/python/dictionaries_vs_lists.jpg)

Como ejemplo, crearemos un diccionario para traducir las palabras en ingl√©s al espa√±ol. Para este diccionario, las claves son cadenas y los valores tambi√©n ser√°n cadenas.

Una forma de crear un diccionario es comenzar con el diccionario vac√≠o y agregar pares clave-valor. **El diccionario vac√≠o se denota `{}`**

````python
eng2sp = {}
eng2sp['one'] = 'uno'
eng2sp['two'] = 'dos'
eng2sp['three'] = 'tres'

print(eng2sp)
````

La primera asignaci√≥n crea un diccionario vac√≠o llamado `eng2sp`. Las otras asignaciones agregan nuevos pares clave-valor al diccionario. El lado izquierdo da el diccionario y la clave que se asocia. El lado derecho da el valor que se asocia con esa clave. Podemos imprimir el valor actual del diccionario de la forma habitual.


Los pares clave-valor del diccionario est√°n separados por comas. Cada par contiene una clave y un valor separados por dos puntos.


El orden de las parejas puede no ser lo que usted esperaba. Python utiliza algoritmos complejos, dise√±ados para un acceso muy r√°pido, para determinar d√≥nde se almacenan los pares clave-valor en un diccionario. Para nuestros prop√≥sitos, podemos pensar que este ordenamiento es impredecible.


Otra forma de crear un diccionario es proporcionar una lista de pares clave-valor utilizando la misma sintaxis que la salida anterior.

````python
eng2sp = {'three': 'tres', 'one': 'uno', 'two': 'dos'}

print(eng2sp)
````

No importa en qu√© orden escribamos los pares. Se accede a los valores de un diccionario con claves, no con √≠ndices, por lo que no hay necesidad de preocuparse por el orden.


Ahora un ejemplo de c√≥mo usamos una clave para buscar el valor correspondiente:

````python
eng2sp = {'three': 'tres', 'one': 'uno', 'two': 'dos'}

print(eng2sp['three'])
````

````python
# Intercambiar clave y valor en un diccionario: Dado un diccionario, intercambia las claves y valores.
def intercambiar_clave_valor(diccionario):
  return {valor: clave for clave, valor in diccionario.items()}

mi_diccionario = {"a": 1, "b": 2, "c": 3}
diccionario_intercambiado = intercambiar_clave_valor(mi_diccionario)
print(f"Diccionario original: {mi_diccionario}")
print(f"Diccionario intercambiado: {diccionario_intercambiado}")
````

* **Encontrar palabras pal√≠ndromas en una lista**: Dada una lista de palabras, devuelve una nueva lista con las palabras que sean pal√≠ndromos.

````python
# Encontrar palabras pal√≠ndromas en una lista: Dada una lista de palabras, devuelve una nueva lista con las palabras que sean pal√≠ndromos.
def es_palindromo(palabra):
  return palabra == palabra[::-1]

def encontrar_palindromos(lista_palabras):
  palindromos = []
  for palabra in lista_palabras:
    if es_palindromo(palabra):
      palindromos.append(palabra)
  return palindromos

lista_de_palabras = ["ala", "casa", "reconocer", "sol", "nivel"]
palabras_palindromas = encontrar_palindromos(lista_de_palabras)
print(f"La lista original de palabras es: {lista_de_palabras}")
print(f"Las palabras pal√≠ndromas son: {palabras_palindromas}")
````

* **Crear una lista de tuplas con √≠ndice y valor**: Dada una lista, genera una lista de tuplas donde cada tupla tenga el √≠ndice y el valor correspondiente.

Salida esperada: [(0, "a"), (1, "b"), (2, "c")]

````python
# Crear una lista de tuplas con √≠ndice y valor: Dada una lista, genera una lista de tuplas donde cada tupla tenga el √≠ndice y el valor correspondiente.

# Salida esperada: [(0, "a"), (1, "b"), (2, "c")]
def lista_a_tuplas_con_indice(lista):
  lista_de_tuplas = []
  for indice, valor in enumerate(lista):
    lista_de_tuplas.append((indice, valor))
  return lista_de_tuplas

mi_lista = ["a", "b", "c"]
lista_de_tuplas = lista_a_tuplas_con_indice(mi_lista)
print(f"La lista original es: {mi_lista}")
print(f"La lista de tuplas con √≠ndice y valor es: {lista_de_tuplas}")
````

  * actualizar_estudiante(lista, id, nuevo_nombre, nueva_edad): Modifica los datos de un estudiante. **¬øC√≥mo har√≠as esto?**
  * eliminar_estudiante(lista, id): Elimina un estudiante.

````python
def crear_estudiante(lista, id, nombre, edad):
  lista.append((id, nombre, edad))
  print(f"Estudiante {nombre} agregado.")

def leer_estudiantes(lista):
  print("Lista de estudiantes:")
  if not lista:
    print("No hay estudiantes en la lista.")
  else:
    for estudiante in lista:
      print(f"ID: {estudiante[0]}, Nombre: {estudiante[1]}, Edad: {estudiante[2]}")

def actualizar_estudiante(lista, id, nuevo_nombre, nueva_edad):
  for i, estudiante in enumerate(lista):
    if estudiante[0] == id:
      # Como las tuplas son inmutables, creamos una nueva tupla y reemplazamos la antigua
      lista[i] = (id, nuevo_nombre, nueva_edad)
      print(f"Estudiante con ID {id} actualizado.")
      return
  print(f"No se encontr√≥ estudiante con ID {id}.")


def eliminar_estudiante(lista, id):
  for i, estudiante in enumerate(lista):
    if estudiante[0] == id:
      del lista[i]
      print(f"Estudiante con ID {id} eliminado.")
      return
  print(f"No se encontr√≥ estudiante con ID {id}.")

# Ejemplo de uso del CRUD
lista_estudiantes = []

crear_estudiante(lista_estudiantes, 1, "Ana", 20)
crear_estudiante(lista_estudiantes, 2, "Luis", 22)
leer_estudiantes(lista_estudiantes)

actualizar_estudiante(lista_estudiantes, 1, "Ana Maria", 21)
leer_estudiantes(lista_estudiantes)

eliminar_estudiante(lista_estudiantes, 2)
leer_estudiantes(lista_estudiantes)

eliminar_estudiante(lista_estudiantes, 3) # Intentar eliminar un estudiante que no existe
````
